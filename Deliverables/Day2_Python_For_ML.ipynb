##### Create vectors and matrices using NumPy

```python
import numpy as np
```

##### 2. Create and define two vectors and matrices

```python
# Define and create two vectors
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])

# Define and create two matrices
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])
```

##### **Observations**

- Defining both vectors and matrices uses the same syntax `np.array()`

##### Compute dot products and matrix multiplication

##### 1. Perform vector addition and multiplication

```python
# Vector addition
vector_sum = vector_a + vector_b
print("Vector Addition:", vector_sum)

# Vector multiplication (dot product)
vector_dot_product = np.dot(vector_a, vector_b)
print("Vector Dot Product:", vector_dot_product)
```

```python
# Matrix addition
matrix_sum = matrix_a + matrix_b
print("Matrix Addition:\n", matrix_sum)

# Matrix multiplication
matrix_product = np.dot(matrix_a, matrix_b)
print("Matrix Multiplication:\n", matrix_product)
```

##### **Observations**

- Addition in both vectors and matrices are the same.
- Vector multiplication (Dot Product) is different to Matrix multiplication.
- In vector multiplication, the product is scalar. It can be defined as the sum of the product of each element (e.g. (a â€¢ b) = a<sub>1</sub>b<sub>1</sub> + a<sub>2</sub>b<sub>2</sub> + a<sub>3</sub>b<sub>3</sub>)
- However, in matrix multiplication, is more complex and its product is not scalar, but produces a new matrix.
- To compute for the matrix multiplication:
    - Calculate c<sub>1,1</sub> which is the dot product of Row 1 of a and Column 1 of b
    - Calculate c<sub>1,2</sub> which is the dot product of Row 1 of a and Column 2 of b
    - Calculate c<sub>2,1</sub> which is the dot product of Row 2 of a and Column 1 of b
    - Calculate c<sub>2,2</sub> which is the dot product of Row 2 of a and Column 2 of b

##### Verify shape compatibility (.shape)

##### 1. Shape Compatibility

```python
# Verify shape compatibility for matrix multiplication
matrix_c = np.array([[1, 2, 3], [4, 5, 6]])
matrix_d = np.array([[7, 8], [9, 10], [11, 12]])

print("Matrix C shape:", matrix_c.shape)
print("Matrix D shape:", matrix_d.shape)

if matrix_c.shape[1] == matrix_d.shape[0]:
    print("Matrices are compatible for multiplication.")
else:
    print("Matrices are not compatible for multiplication.")
```

##### 2. Shape Incompatibility

```python
# Reshape matrix_c to make it incompatible with matrix_d
matrix_c_reshaped = matrix_c.reshape(3, 2)
print("Reshaped Matrix C:\n", matrix_c_reshaped)

# Now multiply the reshaped matrix_c with matrix_d
matrix_new_product = np.dot(matrix_c_reshaped, matrix_d)
print("New Matrix Multiplication Result:\n", matrix_new_product)
```

##### **Observations**

- In vector multiplication, the shape of the matrix needs to be verified using `.shape()` to become valid.

*Note: Matrix multiplication is valid if the number of columns of the first matrix is ***equal*** to the number of rows in the second matrix.*

### Hands-on Tasks

##### Compute ***mean*** and ***variance*** manually and with NumPy

##### 1. Import `NumPy`, `Matplotlib` and `Scipy` library

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

##### 2. Using manual computation ***mean*** and ***variance***

```python
# Compute mean and variance of a dataset manually
data = np.array([10, 12, 23, 23, 16, 23, 21, 16])
mean_manual = np.sum(data) / len(data)
variance_manual = np.sum((data - mean_manual) ** 2) / len(data)
print("Manual Mean:", mean_manual)
print("Manual Variance:", variance_manual)
```

##### 2. Using `NumPy` method for ***mean*** and ***variance*** computation

```python
# Compute mean and variance using NumPy functions
mean_numpy = np.mean(data)
variance_numpy = np.var(data)
print("NumPy Mean:", mean_numpy)
print("NumPy Variance:", variance_numpy)
```

##### **Observations**

- ***Mean*** and ***average*** has the same computation which is the sum of all the elements divided by the number of elements. ****mean*** = (a<sub>1</sub> + a<sub>2</sub> + ... + a<sub>n</sub>) / N*
- ***Variance*** can be computed as the sum of the squares of the difference of each data to the ***mean***, and divided by the number of elements.
- ***Variance*** interpretation is:
    - High ***variance*** means data points are spread out far from the mean.
    - Low ***variance*** means data points are clustered close to the mean.
- Difference of ***mean*** and ***variance*** can be defined by the following:
    - ***Mean*** measures the **average value of the dataset** while ***variance*** measures the **average squared distance of each data point** with mean as the reference point.
    - ***Mean*** tells you **where the data is centered** while ***variance*** tells **how far** each data points from the center.


##### Visualize a normal distribution

```python
# Visualize normal distribution
standard_devs = np.std(data)
x = np.linspace(mean_manual - 4*standard_devs, mean_manual + 4*standard_devs, 100)
pdf = norm.pdf(x, mean_manual, standard_devs)
plt.plot(x, pdf, label='Normal Distribution')
plt.title('Normal Distribution of Data')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

##### **Observations**

- ***Standard Deviation*** determines the spread (width) of the curve while ***mean*** determines the center or peak of the curve.
- ***Normal Distribution*** describes the values of a variable are distributed for a population.
- Plot of ***Normal Distribution*** given the data points, is observed clearly using `plt.show()` method. It is a inverse U-shaped graph wherein the highest peak is the ***mean*** of the data. 
- Importance of normal distribution lies into the statistical analysis and machine learning techniques that rely on statistical inference since it is considered as a highly accuracte model.

##### Explain bias vs variance in your own words (markdown)

##### Definition

* Bias - error introduced by a simple model that fails to capture the complexity and underlying patterns of data.
* Variance - error introduced by a complex model that is too sensitive to small changes or varying parameters in the data.

Characteristics:

| | Model Behavior | Performance | Prediction | Causes | Goal
| --- | --- | --- | --- | --- | --- |
| Bias | Underfitting | Performs poorly in Training and Testing datasets | Low Variance | Simple algorithms, few features | Decrease bias by increasing model complexity
| Variance | Overfitting | Performs well in Training dataset but poorly in Testing datasets | High Variance | Complex algorithms, insufficient training data | Decrease variance by simplifying model complexity

In my own words, ***bias*** is generally caused by having overly simplified model which usually means that there are few features to extract to that model. Since there are only few features, it is hard to find underlying patterns in order to create a good prediction. On the other hand, ***variance*** is the opposite of bias, uses a overly complex model, which means more features are extracted resulting into high sensitivity in drastic change in data. The goal of the data scientist is to balance out ***bias*** and ***variance*** by computing the total error which can be defined as the **sum of the square of the bias, variance and irreducible errors**.

## Gradient Descent Implementation

### Objectives

- Learn how models actually learn
- Understand optimization mechanics
- Prepare foundation for neural networks

### Hands-on Tasks

##### Implement Gradient Descent step-by-step

##### 1. Prepare the components for implementing gradient descent.

- Model (Use Linear Regression model earlier)
- Loss function (Use MSE earlier)
- The Gradient (to be determined)

##### 2. Initialize the model parameters(slope and y-intercept) to small random values or zero.

##### 3. Choose a learning rate (default = 0.01)

##### 4. Set number of iterations (epochs=1000) or a convergence threshold

##### 5. Calculate Predictions, Loss, Gradients and update parameters.

```python
def gradient_descent_fixed(X, y, learning_rate=0.01, epochs=1000):
    m = len(y) # sample size

    # 1. Reshape y to ensure it's a column vector (m, 1)
    if y.ndim == 1:
        y = y.reshape(-1, 1) 
    
    # Initialize theta (parameters) as a column vector (n_features, 1)
    # X.shape[1] is the number of features (2 in X_b)
    theta = np.zeros((X.shape[1], 1))
    
    loss_history = []
    
    for _ in range(epochs):
        # Predictions: X_b @ theta will result in (m, 1)
        predictions = X @ theta
        errors = predictions - y
        
        # Calculate Loss (MSE)
        loss = np.mean(errors ** 2)
        loss_history.append(loss)
        
        # Gradients: X.T @ errors will result in (n_features, 1)
        # @ is the matrix multiplication operator in Python 3.5+
        gradients = (2/m) * X.T @ errors
        
        # 4. Parameter Update: Now both theta and gradients have shape (2, 1),
        # making the subtraction possible via broadcasting with the scalar alpha.
        theta -= learning_rate * gradients
        
    return theta.flatten(), loss_history # Return flattened theta for easier printing

# Prepare data for gradient descent
# X_b is (100, 2)
X_b = np.c_[np.ones((X.shape[0], 1)), X]

# Run Gradient Descent
theta_gd, loss_history = gradient_descent_fixed(X_b, y, learning_rate=0.01, epochs=1000)

print("Parameters from Gradient Descent (Intercept, Slope):", theta_gd)
print("Mean Squared Error after Gradient Descent:", loss_history[-1])
```

##### Plot loss vs iterations

```python
# --- PLOT LOSS VS. ITERATIONS ---
plt.figure(figsize=(10, 6))
# Create the x-axis (iterations)
iterations = range(len(loss_history)) 
plt.plot(iterations, loss_history, color='purple')
plt.title('Loss (MSE) vs. Iterations (Training Progress)')
plt.xlabel('Iteration')
plt.ylabel('Loss (Mean Squared Error)')
plt.grid(True)
plt.show()
```

## Interview Readiness Tip (Important)

##### By the end of Day 2, you should be able to clearly explain:

- How linear regression works mathematically
    - Linear Regression works as a method to describe and quantify the relationship between two variables (in machine learning context, most likely between a feature and an output). Furthermore, this also tells the best fit for the dataset which entails that this line minimizes the residuals from each data point to that specific line. The importance of linear regression is you would be able to closely predict the output given the feature (represented by the x-axis).

- Why gradient descent converges (or fails)
    - Gradient descent converges or diverges is dependent on learning rate parameter. Learning rate is the variable that is multiplied to update the gradient in the iteration loop while the epochs is described as the number of iterations of the gradient descent function. A model is considered converging when loss curve is smoothly decreasing and eventuall flattening out or reaching an asymptote. On the other hand, a diverging model occurs when the loss curve either increases with each iteration or oscillates violently.

- What learning rate controls
    - Learning rate controls the step of the gradient descent algorithm should take. A higher learning rate tends to overshoot or diverge the model whereas lower learning rate tends to demand more iterations to reach the minimum. It is important to determine the right learning rate to stabilize the parameter as well as reaching the minimum in the cost function quickly with lesser iterations.

- Bias vs variance tradeoff
    - As described earlier, bias and variance depends on the model or algorithm used. Having a higher bias and low variance usually happens in oversimplified models while overly complex models tends to have lower bias and higher variance. Alleviating higher bias by utilizing a more complex algorithm or adding polynomial features can be done while simplifying complex models such as the application of regularization and feature reductions can be done to minimize higher variance. It is important to do necessary adjustments to balance these two for the model to provide correct prediction and results.
